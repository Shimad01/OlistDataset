```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
review  = pd.read_csv('../datasets/olist_order_reviews_dataset.csv')
item    = pd.read_csv('../datasets/olist_order_items_dataset.csv')
product = pd.read_csv('../datasets/olist_products_dataset.csv')

review.head(3)
```

```python
item.head(3)
```

```python
product.head(3)
```

```python
review_pre = review.dropna(subset = ['review_comment_message'])
review_pre.info()
```

```python
df = review_pre.merge(item, how = 'left', on = 'order_id').merge(product, how = 'left', on = 'product_id')
df.head()
```

```python
df.drop_duplicates(subset = ['order_id','product_category_name'], inplace = True)
df.head()
```

```python
df.drop(labels = ['review_id', 'review_creation_date','review_answer_timestamp', 'order_item_id', 'seller_id',
       'shipping_limit_date', 'product_name_lenght','product_description_lenght', 'product_photos_qty', 'product_weight_g',
       'product_length_cm', 'product_height_cm', 'product_width_cm'], axis = 1, inplace = True)
df.head(3)
```

```python
csat = {1:0,2:0,3:0,4:1,5:1}

df.review_score = df.review_score.map(csat)
df.head(3)
```
